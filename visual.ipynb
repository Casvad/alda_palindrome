```python
with open("out.txt", "r") as file:
    data = file.read()
    data = data.replace("[", "").replace("]", "").split("\n")
    data.pop()
data = [list(map(int, line.strip("").split(", "))) for line in data]
```

```python
data
```

```python
import matplotlib.pyplot as plt

iterative = [item[1] for item in data]
recursive = [item[2] for item in data]
reverse = [item[3] for item in data]
time_values = [item[0] for item in data]

plt.figure(figsize=(10, 6))
plt.plot(time_values, iterative, label="Iterative", marker="o")
plt.plot(time_values, recursive, label="Recursive", marker="o")
plt.plot(time_values, reverse, label="Reverse", marker="o")


plt.xlabel("Dataset size")
plt.ylabel("Time (s)")
plt.title("Sorting comparison")
plt.legend()
plt.grid(True)
plt.show()
```
